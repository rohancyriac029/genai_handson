```python
#PES2UG23CS490
from groq import Groq
from dotenv import load_dotenv
import os

load_dotenv()

client = Groq(api_key=os.environ.get("GROQ_API_KEY"))

MODEL_NAME = "llama-3.1-8b-instant"
```

```python
#PES2UG23CS490
MODEL_CONFIG = {
    "technical": {
        "system_prompt": """You are a Senior Technical Support Engineer.
Be rigorous, precise, and code-focused.
If debugging, explain the issue clearly and provide corrected code.
Avoid unnecessary fluff."""
    },

    "billing": {
        "system_prompt": """You are a Billing and Subscription Specialist.
Be empathetic and professional.
Focus on refund policies, charges, invoices, and subscription management.
Provide clear next steps for resolution."""
    },

    "general": {
        "system_prompt": """You are a helpful and friendly general assistant.
Answer conversational queries clearly and naturally."""
    },

    # BONUS TOOL EXPERT
    "tool": {
        "system_prompt": """You are a financial data assistant.
When necessary, you may fetch real-time financial data using tools."""
    }
}
```

```python
#PES2UG23CS490
def route_prompt(user_input):
    router_prompt = f"""
Classify the following user queries into one of these categories:
[technical, billing, general, tool]

Return ONLY the category name. No explanation.

User Query:
{user_input}
"""

    response = client.chat.completions.create(
        model=MODEL_NAME,
        temperature=0,
        messages=[
            {"role": "system", "content": "You are a strict classifier."},
            {"role": "user", "content": router_prompt}
        ]
    )

    category = response.choices[0].message.content.strip().lower()
    return category
```

```python
#PES2UG23CS490
import random

def get_bitcoin_price():
    """
    Mock function to simulate fetching Bitcoin price.
    """
    mock_price = random.randint(30000, 70000)
    return f"The current price of Bitcoin is approximately ${mock_price} USD."
```

```python
#PES2UG23CS490
def process_request(user_input):

    category = route_prompt(user_input)
    print(f"[Router Decision]: {category}")

    if category == "tool":
        if "bitcoin" in user_input.lower():
            return get_bitcoin_price()
        else:
            return "Tool requested but no matching tool available."


    system_prompt = MODEL_CONFIG.get(category, MODEL_CONFIG["general"])["system_prompt"]


    response = client.chat.completions.create(
        model=MODEL_NAME,
        temperature=0.7,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_input}
        ]
    )

    return response.choices[0].message.content
```

```python
#PES2UG23CS490
query1 = "My python script is throwing an IndexError on line 5."

response1 = process_request(query1)
print("\nFinal Response:\n")
print(response1)
```
